```python
def dist(p,q):
    return np.sqrt(np.sum((p - q) ** 2))
```

```python
# Here K is our Hyper Parameter
```

```python
def knn_c(X,y,xt,k=5):
    m = X.shape[0]
    dlist = []
    for i in range(m):
        d = dist(X[i],xt)
        dlist.append((d,y[i]))

    dlist = np.array(sorted(dlist)[:k])

    labels = dlist[:,1]

    labels , cnts = np.unique(labels,return_counts = True)
    idx = cnts.argmax()

    pred = labels[idx]
    return int(pred)
        
```

```python
# KNN for Regression

def knn_r(X,y,xt,k=5):
    m = X.shape[0]
    dlist = []
    for i in range(m):
        d = dist(X[i],xt)
        dlist.append((d,y[i]))

    # Sort by distance and select k-nearest
    dlist = sorted(dlist, key=lambda x: x[0])[:k]
    
    # Extract labels and compute mean prediction
    pred = np.mean([val[1] for val in dlist])
    return pred
        
```
